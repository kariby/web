```python
import pandas as pd

def toStr(i):
    if (i < 10):
        return '0' + str(i)
    return str(i);

def maxDays(i):
    if (i == 1 or i == 3 or i == 5 or i == 7 or i == 8 or i == 10 or i == 12):
        return 31
    elif (i == 4 or i == 6 or i == 9 or i == 11):
        return 30
    else:
        return 28
        
frames = []
for i in range(1, 12):
    for j in range(1, maxDays(i)):
        frame = pd.read_csv('./data20190320035610/box2018' + toStr(i) + toStr(j) + 'qmin.min'  , comment='#')
        frames.append(frame)
```

```python
hdf5store = pd.HDFStore('mydata.h5')
```

```python
%ls mydata.h5
```

```python
for i in frames:
    hdf5store[str(i)] = i
```
